```python
# Anulaciones
file_tab = DATA_PATH / "df_muestra_vars_eng.csv"
df_muestra_vars_eng = pd.read_csv(file_tab)

print(df_muestra_vars_eng.shape)
df_muestra_vars_eng.head(3)
```

# Variables

```python
var_target = 'anula'
vars_key=['id_poliza']


vars_flag=['anul_flag_gestion_multas', # gestion_multas_
           'anul_flag_vehiculo_sustitucion', #Viene de garantia_ampliada_norm
           'anul_flag_tipo_pago_anual'] #Viene de tipo_pago_norm


vars_num_socio = [c for c in  df_muestra_vars_eng.columns if c.startswith("socio_")]
vars_num_anul = [c for c in  df_muestra_vars_eng.columns if c.startswith("anul_")]

vars_total=vars_key+[var_target]+vars_num_socio+vars_num_anul

assert len(vars_total)==len(df_muestra_vars_eng.columns), "Faltan variables"
```

# Split train-val-test

```python
df_muestra_vars_eng[vars_total].head()
df_muestra_vars_eng.rename(columns={'anula': 'target'}, inplace=True)

if "anula" in vars_total:
    idx = vars_total.index("anula")
    vars_total[idx] = "target"
```

```python
assert df_muestra_vars_eng.index.name == "_row_id"
```

```python
# Copia + índice estable
df = df_muestra_vars_eng.copy()
assert {"target", "id_poliza"} <= set(df.columns)

df = df.reset_index(drop=True)                    # limpia índice previo
df["_row_id"] = np.arange(len(df), dtype=int)     # índice estable por fila
df = df.set_index("_row_id")                      # usarlo como index

# --- 60/20/20 estratificado ---
df_train, df_temp = train_test_split(
    df, test_size=0.40, stratify=df["target"], random_state=42
)
df_test, df_val = train_test_split(
    df_temp, test_size=0.50, stratify=df_temp["target"], random_state=42
)

# Features: quitar id_poliza para entrenar
X_train = df_train.drop(columns=["target", "id_poliza"])
y_train = df_train["target"]

X_val   = df_val.drop(columns=["target", "id_poliza"])
y_val   = df_val["target"]

X_test  = df_test.drop(columns=["target", "id_poliza"])
y_test  = df_test["target"]

# (Opcional) guardar un "sidecar" con las llaves para trazabilidad
keys_train = df_train[["id_poliza"]]
keys_val   = df_val[["id_poliza"]]
keys_test  = df_test[["id_poliza"]]

# --- Guardado: mantener el índice para poder alinear por index al leer ---
X_train.to_csv( DATA_PATH / "X_train.csv", index=True, index_label="_row_id")
y_train.to_csv( DATA_PATH / "y_train.csv", index=True, index_label="_row_id")
keys_train.to_csv( DATA_PATH / "keys_train.csv", index=True, index_label="_row_id")

X_val.to_csv( DATA_PATH / "X_val.csv", index=True, index_label="_row_id")
y_val.to_csv( DATA_PATH / "y_val.csv", index=True, index_label="_row_id")
keys_val.to_csv( DATA_PATH / "keys_val.csv", index=True, index_label="_row_id")

X_test.to_csv( DATA_PATH / "X_test.csv", index=True, index_label="_row_id")
y_test.to_csv( DATA_PATH / "y_test.csv", index=True, index_label="_row_id")
keys_test.to_csv( DATA_PATH / "keys_test.csv", index=True, index_label="_row_id")
```
